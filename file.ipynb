#### 2. K-Means Clustering on a Multi-Class and Multi-Label Data Set

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import calinski_harabaz_score
```

#### Creating separate dataframes for train and dividing the labels into three different dataframes namely Family,SPecies and Genus

```python
df=pd.read_csv("Frogs_MFCCs.csv")
df.drop('RecordID', axis=1, inplace=True)
df.head()
y=df[["Family","Genus","Species"]]
Family=y[["Family"]]
Genus=y[["Genus"]]
Species=y[["Species"]]
X=df.drop(["Family","Genus","Species"],axis=1)
X.head()
```

#### (a) Use k-means clustering on Anuran Calls (MFCCs) Data Set of Homework 4. Choose k automatically based on one of the methods provided in the slides (CH or Gap Statistics or scree plots) or any other method you know.

We observe that the number of clusters is 4 by using silhouette score


```python
from sklearn.metrics import silhouette_score
num_clusters=np.arange(2,10)
results={}
for size in num_clusters:
    model=KMeans(n_clusters=size).fit(X)
    predictions=model.predict(X)
    results[size]=silhouette_score(X,predictions)
best_size=max(results,key=results.get)
print(best_size)
```

#### (b) In each cluster, determine which family is the majority by reading the true labels. Repeat for genus and species. 

```python

kmeans=KMeans(n_clusters=4).fit(X)
clusters=kmeans.labels_


index0 = np.argwhere(clusters == 0)
index1 = np.argwhere(clusters == 1)
index2 = np.argwhere(clusters == 2)
index3 = np.argwhere(clusters == 3)


family1=pd.DataFrame(columns=['Family'])
g1=pd.DataFrame(columns=['Genus'])
s1=pd.DataFrame(columns=['Species'])

family2=pd.DataFrame(columns=['Family'])
g2=pd.DataFrame(columns=['Genus'])
s2=pd.DataFrame(columns=['Species'])

family3=pd.DataFrame(columns=['Family'])
g3=pd.DataFrame(columns=['Genus'])
s3=pd.DataFrame(columns=['Species'])

family4=pd.DataFrame(columns=['Family'])
g4=pd.DataFrame(columns=['Genus'])
s4=pd.DataFrame(columns=['Species'])
```

####  2(c) Now for each cluster you have a majority label triplet (family, genus, species). Calculate the average Hamming distance between the true labels and the labels assigned by clusters.

#### Hamming score for each of family,species and label is found out by using the predictions from the Kmeans clustering and the true labels.

```python
from sklearn.metrics import hamming_loss
pred=kmeans.predict(X)
pf=pred
pg=pred
ps=pre

new_family=pd.DataFrame(pf,columns=Family.columns)
new_genus=pd.DataFrame(pg,columns=Genus.columns)
new_species=pd.DataFrame(ps,columns=Species.columns)

new_family=new_family.replace((0,1,2,3),('Leptodactylidae ',' Leptodactylidae','Hylidae ','Leptodactylidae '))
new_genus=new_genus.replace((0,1,2,3),('Adenomera ','Adenomera ','Hypsiboas','Hypsiboas'))
new_species=new_species.replace((0,1,2,3),('AdenomeraHylaedactylus ','AdenomeraHylaedactylus','HypsiboasCordobae','HypsiboasCinerascens'))

family_hscore=hamming_loss(Family,new_family)
print("hamming score for family is")
print(family_hscore)

genus_hscore=hamming_loss(Genus,new_genus)
print("hamming score for genus is")
print(genus_hscore)

species_hscore=hamming_loss(Species,new_species)
print("hamming score for species is")
print(species_hscore)

```
